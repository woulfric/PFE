```python
import pandas as pd
import numpy as np

def sequential_Data_per_mission(Data, sequence_length, mission_length=64):
    X = []
    y = []
    
    num_missions = len(Data) // mission_length
    
    for mission_index in range(num_missions):
        start_index = mission_index * mission_length
        end_index = start_index + mission_length
        
        mission_data = Data.iloc[start_index:end_index]
        
        for i in range(len(mission_data) - sequence_length):
            # Select the sequence for features and corresponding label
            sequence_x = mission_data.iloc[i:i+sequence_length, :-1]  # Exclude the label column
            sequence_y = mission_data.iloc[i+sequence_length, -1]     # Label corresponding to the sequence end
            
            sequence_x = sequence_x.to_numpy()
            sequence_y = np.array([sequence_y])
            
            X.append(sequence_x)
            y.append(sequence_y)
    
    # Convert lists to NumPy arrays
    X = np.array(X)
    y = np.array(y)
    
    return X, y


sequence_length = 4

true_temp_data = true_missions
# true_temp_data = Min_Max_true
# true_temp_data = std_true

false_temp_data = false_missions
# false_temp_data = Min_Max_false
# false_temp_data = std_false

X_true, y_true = sequential_Data_per_mission(true_temp_data, sequence_length)
X_false, y_false = sequential_Data_per_mission(false_temp_data, sequence_length)


# VÃ©rification des dimensions
print(f"Forme de X_true : {X_true.shape}")
print(f"Forme de y_true : {y_true.shape}")
# VÃ©rification des dimensions
print(f"Forme de X_false : {X_false.shape}")
print(f"Forme de y_false : {y_false.shape}")

X = np.concatenate((X_true, X_false))
y = np.concatenate((y_true, y_false))
print("______________________________")
print(f"Forme de X : {X.shape}")
print(f"Forme de y : {y.shape}")
```

```python
import numpy as np

# Create indices array
indices = np.arange(X.shape[0])

# Shuffle the indices
np.random.shuffle(indices)

# Use the shuffled indices to shuffle both data and labels arrays
X = X[indices]
y = y[indices]



print("Shuffled data:")
print(X)
print("Shuffled labels:")
print(y)
```

## **Model**

### **Model 1**

```python
import time
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, TimeDistributed, BatchNormalization
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping

x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=30)
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

model = Sequential([
    LSTM(units=14, input_shape=(x_train.shape[1], x_train.shape[2]), return_sequences=True),
    BatchNormalization(),
    Dropout(0.3),
    LSTM(units=7, return_sequences=False),
    Dropout(0.3),
    BatchNormalization(),
    Dense(units=1, activation='sigmoid')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

sart_time = time.time()
history = model.fit(x_train, y_train, epochs=100, batch_size=32, validation_split=0.2, callbacks=[early_stopping])
# history = model.fit(x_train, y_train, epochs=100, batch_size=128, validation_split=0.2)
end_time = time.time()


loss, accuracy = model.evaluate(x_test, y_test)
print("Training time : ",end_time - sart_time)
```

### **Model 3**

```python
import time
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping

x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=30)
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

model = Sequential([
    LSTM(units=512, input_shape=(x_train.shape[1], x_train.shape[2]), return_sequences=True),
    BatchNormalization(),
    Dropout(0.3),
    LSTM(units=128, return_sequences=True),
    BatchNormalization(),
    Dropout(0.3),
    LSTM(units=64, return_sequences=False),
    BatchNormalization(),
    Dropout(0.3),
    Dense(units=64),
    BatchNormalization(),
    Dense(units=1, activation='sigmoid')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', 'Precision', 'Recall'])

sart_time = time.time()
history = model.fit(x_train, y_train, epochs=100, batch_size=32, validation_split=0.2, callbacks=[early_stopping])
# history = model.fit(x_train, y_train, epochs=100, batch_size=128, validation_split=0.2)
end_time = time.time()


loss, accuracy, precision, recall = model.evaluate(x_test, y_test)
print("Training time : ",end_time - sart_time)
print(f'Loss: {loss}, Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}')
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Metrics')
plt.xlabel('Epoch')
plt.ylabel('Metrics')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Making predictions
y_pred_prob = model.predict(x_test)
y_pred = (y_pred_prob > 0.5).astype(int)

# Reshape y_test and y_pred if they are sequences
y_test_reshaped = y_test.reshape(-1, 1)
y_pred_reshaped = y_pred.reshape(-1, 1)

# Compute confusion matrix
conf_matrix = confusion_matrix(y_test_reshaped, y_pred_reshaped)

# Print confusion matrix
print("Confusion Matrix:")
print(conf_matrix)

# Optionally, visualize the confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Print classification report for detailed metrics
print("Classification Report:")
print(classification_report(y_test_reshaped, y_pred_reshaped))
```
