### **LSTM**

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, TimeDistributed, BatchNormalization
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping

# x = data.drop('label', axis=1)
# x = data_MinMax.drop('label', axis=1)
x = data_std.drop('label', axis=1)
y = data['label']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=30)
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

x_train_reshaped = np.expand_dims(x_train, axis=1)
x_test_reshaped = np.expand_dims(x_test, axis=1)


model = Sequential([
    LSTM(units=64, input_shape=(1,14), return_sequences=True),
    BatchNormalization(),
    Dropout(0.2),
    LSTM(units=32, return_sequences=True),
    BatchNormalization(),
    Dense(units=1, activation='sigmoid')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

sart_time = time.time()
history = model.fit(x_train_reshaped, y_train, epochs=100, batch_size=128, validation_split=0.2, callbacks=[early_stopping])
end_time = time.time()


loss, accuracy = model.evaluate(x_test_reshaped, y_test)
print("Training time : ",end_time - sart_time)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.layers import BatchNormalization
from sklearn.model_selection import train_test_split

# Define the generator network
def make_generator_model(latent_dim, num_features):
    model = models.Sequential([
        layers.Dense(128, input_dim=latent_dim, activation='relu'),
        layers.Dense(256, activation='relu'),
        layers.Dense(num_features, activation='tanh')
    ])
    return model

# Define the discriminator network
def make_discriminator_model(num_features):
    model = models.Sequential([
        layers.Dense(128, input_shape=(num_features,), activation='relu'),
        BatchNormalization(),
        layers.Dropout(0.2),
        layers.Dense(64, activation='relu'),
        BatchNormalization(),
        layers.Dense(1, activation='sigmoid')
    ])
    return model

# Define the GAN model
def make_gan_model(generator, discriminator):
    discriminator.trainable = False
    gan_input = layers.Input(shape=(latent_dim,))
    gan_output = discriminator(generator(gan_input))
    gan = models.Model(gan_input, gan_output)
    gan.compile(loss='binary_crossentropy', optimizer='adam')
    return gan

# Define training parameters
latent_dim = 100  # Dimensionality of the latent space
num_features = 14  # Number of features in the dataset
num_epochs = 10
batch_size = 32
toggle_interval = 2  # Interval (in epochs) to re-enable discriminator training

# Create instances of generator and discriminator
generator = make_generator_model(latent_dim, num_features)
discriminator = make_discriminator_model(num_features)

discriminator.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Create GAN model
gan = make_gan_model(generator, discriminator)

# Assume true_data is already loaded and preprocessed
X_train = true_MinMax.drop('label', axis=1)
y_train = true_data['label']

# Training loop
for epoch in range(num_epochs):
    # Shuffle the training data
    indices = np.arange(X_train.shape[0])
    np.random.shuffle(indices)
    X_train = X_train.iloc[indices]

    if epoch % toggle_interval == 0:
        discriminator.trainable = True
        discriminator.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    for i in range(len(X_train) // batch_size):
        # Train discriminator on real data
        real_data = X_train.iloc[i * batch_size: (i + 1) * batch_size]
        real_labels = np.ones((batch_size, 1))
        d_loss_real, d_acc_real = discriminator.train_on_batch(real_data, real_labels)

        # Generate fake data
        noise = np.random.normal(0, 1, size=(batch_size, latent_dim))
        fake_data = generator.predict(noise)
        fake_labels = np.zeros((batch_size, 1))

        # Train discriminator on fake data
        d_loss_fake, d_acc_fake = discriminator.train_on_batch(fake_data, fake_labels)

        # Disable discriminator training before training GAN
        discriminator.trainable = False

        # Train generator
        noise = np.random.normal(0, 1, size=(batch_size, latent_dim))
        valid_labels = np.ones((batch_size, 1))
        g_loss = gan.train_on_batch(noise, valid_labels)

        # Print progress
        if i % 100 == 0:
            print(f"Epoch {epoch}/{num_epochs}, Batch {i}/{len(X_train) // batch_size}")
            print(f"D Loss Real: {d_loss_real:.4f}, D Acc Real: {d_acc_real:.4f}")
            print(f"D Loss Fake: {d_loss_fake:.4f}, D Acc Fake: {d_acc_fake:.4f}")
            print(f"G Loss: {g_loss:.4f}")

    # Optionally re-enable discriminator training at the end of the epoch
    if epoch % toggle_interval == 0:
        discriminator.trainable = True
        discriminator.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Generate synthetic data
num_samples = 1000
noise = np.random.normal(0, 1, size=(num_samples, latent_dim))
generated_data = generator.predict(noise)

# Save or use the generated data for further analysis
# np.savetxt("generated_data.csv", generated_data, delimiter=",")
```
