# **Prediction LSTM**

```python
import pandas as pd
import numpy as np
import time
from tensorflow.keras.models import load_model
from sklearn.preprocessing import StandardScaler

test_model = load_model("Models\Lstm_Many_to_one_1.h5")
# test_model = model

def prediction(mission, point=64):

    scaler = StandardScaler()
    mission_std = scaler.fit_transform(mission)
    mission_std = pd.DataFrame(mission_std, columns=mission.columns)

    # data = mission
    data = mission_std
    # data = mission

    sequence_length = 4
    sequences = []
    for i in range(0, len(data) - sequence_length + 1):
        sequence = data.iloc[i:i + sequence_length].values
        sequences.append(sequence)

    X = np.array(sequences)

    print(X.shape)
    
    sart_time = time.time()
    result = test_model.predict(X)
    end_time = time.time()
    print("Prediction result: ", (result[0:point] > 0.5).astype(int))
    # print("Prediction result: ", result)
    print("prediction time : ",end_time-sart_time)
```

```python
mission = test_true[:]
# mission = test_false[:]


prediction(mission, 30)
```
