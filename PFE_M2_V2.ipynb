```python
true_label_1=[1 for index in range(0,true_data_1.shape[0])]
true_label_2=[1 for index in range(0,true_data_2.shape[0])]
false_label_1=[0 for index in range(0,false_data_1.shape[0])]
false_label_2=[0 for index in range(0,false_data_2.shape[0])]
print(len(true_label_1))
print(len(true_label_2))
print(len(false_label_1))
print(len(false_label_2))
```

#### **Without Normalisation**

```python
from sklearn.model_selection import train_test_split

x = data.drop('label', axis=1)
y = data['label']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=30)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=11)

sart_time = time.time()
knn.fit(x_train,y_train)
end_time = time.time()
print("Training time : ",end_time - sart_time)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
y_pred=knn.predict(x_test)

print("Train Score : ", knn.score(x_train,y_train))
print("Test Score : ", knn.score(x_test,y_test))
print("Precision : ",precision_score(y_pred,y_test))
print("Recall : ",recall_score(y_pred,y_test))
print("F1 Score : ",f1_score(y_pred,y_test))
print("Confusion Matrix : ")
print(confusion_matrix(y_test, y_pred))
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, TimeDistributed, BatchNormalization
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping

# x = data.drop('label', axis=1)
# x = data_MinMax.drop('label', axis=1)
x = data_std.drop('label', axis=1)
y = data['label']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=30)
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

x_train_reshaped = np.expand_dims(x_train, axis=1)
x_test_reshaped = np.expand_dims(x_test, axis=1)


model = Sequential([
    LSTM(units=64, input_shape=(1,14), return_sequences=True),
    BatchNormalization(),
    Dropout(0.2),
    LSTM(units=32, return_sequences=True),
    BatchNormalization(),
    Dense(units=1, activation='sigmoid')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

sart_time = time.time()
history = model.fit(x_train_reshaped, y_train, epochs=100, batch_size=128, validation_split=0.2, callbacks=[early_stopping])
end_time = time.time()


loss, accuracy = model.evaluate(x_test_reshaped, y_test)
print("Training time : ",end_time - sart_time)
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, TimeDistributed, BatchNormalization
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping

# x = data.drop('label', axis=1)
# x = data_MinMax.drop('label', axis=1)
x = data_std.drop('label', axis=1)
y = data['label']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=30)
early_stopping = EarlyStopping(monitor='val_accuracy', patience=5, restore_best_weights=True)

x_train_reshaped = np.expand_dims(x_train, axis=1)
x_test_reshaped = np.expand_dims(x_test, axis=1)


model = Sequential([
    LSTM(units=256, input_shape=(1,14), return_sequences=True),
    BatchNormalization(),
    Dropout(0.2),
    LSTM(units=512, return_sequences=True),
    BatchNormalization(),
    LSTM(units=1024, return_sequences=True),
    BatchNormalization(),
    Dropout(0.2),
    Dense(units=1, activation='sigmoid')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

sart_time = time.time()
history = model.fit(x_train_reshaped, y_train, epochs=100, batch_size=128, validation_split=0.2, callbacks=[early_stopping])
end_time = time.time()


loss, accuracy = model.evaluate(x_test_reshaped, y_test)
print("Training time : ",end_time - sart_time)
```

**MLP With Grid Search**

```python
import time
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, InputLayer, BatchNormalization, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam 
from sklearn.model_selection import train_test_split

# x = data_MinMax.drop('label', axis=1)
x = data_std.drop('label', axis=1)
y = data['label']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=30)

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=0.0001)



def create_model(units1=256, units2=128, units3=64, learning_rate=0.001, dropout_rate=0.2):
    model = Sequential([
        InputLayer(input_shape=(x_train.shape[1],)),
        BatchNormalization(),
        Dense(units=units1, activation='relu'),
        Dropout(dropout_rate),
        BatchNormalization(),
        Dense(units=units2, activation='relu'),
        BatchNormalization(),
        Dense(units=units3, activation='relu'),
        BatchNormalization(),
        Dense(units=1, activation='sigmoid')
    ])
    
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return model

param_grid = {
    'units1': [128, 256],
    'units2': [64, 128],
    'units3': [32, 64],
    'learning_rate': [0.001, 0.01],
    'dropout_rate': [0.2, 0.3],
    'batch_size': [32, 64],
    'epochs': [50, 100]
}


# Create the KerasClassifier wrapper
model = KerasClassifier(build_fn=create_model, verbose=0)

# Create GridSearchCV object
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, verbose=2, n_jobs=-1)

# Fit the grid search
grid_result = grid.fit(x_train, y_train)

# Print the best parameters and best score
print(f"Best: {grid_result.best_score_} using {grid_result.best_params_}")


# Get the best model
best_model = grid_result.best_estimator_.model

# Evaluate the best model on the test set
loss, accuracy = best_model.evaluate(x_test, y_test)
print(f'Loss: {loss}, Accuracy: {accuracy}')
```
